```python
# Create empty dictionary to capture our output
song_dictionary = {}

# Loop over each song dictionary within our songs list
for song in first_twenty_songs:
    # Grabbing track name - will be our key
    track_name = song['name']
    
    # Creating an empty list to capture our artists
    artist_list = []
    # Looping over our artists because sometimes more than 1
    for artist in song['artists']:
        # Adding each artist's name to our artist list
        artist_list.append(artist['name'])
        
    # Adds a new key/value pair to our dictionary
    # Key is the track name, value is the tuple version of our artist list
    song_dictionary[track_name] = tuple(artist_list)
    
song_dictionary
```

## Task 15: Write a function

Create a function with takes an **artist name** and the **song_dictionary** as arguments, and returns a `list` of songs written by that artist. 

```python
call_on_students(1)
```

```python
# Easier to do things outside of a function first
# Let's try for Giveon
for key, value in song_dictionary.items():

    if 'The Weeknd' in value:
        print(key)
```

```python
# Your code here

def find_song_by_artist(artist_name, song_dict):
    
    '''
    Parameters:
    artist_name: a string of an artist's name to be used to search the dictionary
    song_dict:  a dictionary of top_twenty songs with song name as keys and a list of 
    artist names as values
    
    Returns:
    A list of songs which the given artist appeared on
    '''
    # Creating an empty list to populate with our songs
    song_list = []
    # Loop over our dictionary of songs
    for key, value in song_dict.items():
        # Check if the artist_name argument is in the values of the dict
        if artist_name in value:
            # If so, append the key (song name) to the song_list
            song_list.append(key)
            
    return song_list
```
